### Generating Code to simulate the total avarage range a vehicle will obtain for both traditional and electric vehicle.

```python
# Code for Traditional vehicles
# Define standard fuel tank capacities (in liters) for petrol and diesel vehicles
petrol_tank_capacity = 40  # Adjust based on typical petrol vehicle models
diesel_tank_capacity = 50  # Adjust based on typical diesel vehicle models

# Function to calculate average range based on fuel consumption
def calculate_average_range(row):
    # Extract fuel type and fuel consumption
    fuel_type = row['fuel_type']
    fuel_consumption = row['fuel_consumption_l_100km']

    # Determine the appropriate tank capacity based on fuel type
    if fuel_type.lower() == 'petrol':
        tank_capacity = petrol_tank_capacity
    elif fuel_type.lower() == 'diesel':
        tank_capacity = diesel_tank_capacity
    else:
        return None  # Return None for non-petrol or non-diesel vehicles

    # Ensure the fuel consumption is a float
    if isinstance(fuel_consumption, str):
        # Convert fuel consumption from string to float (e.g., "4,6 l/100 km" to 4.6)
        try:
            fuel_consumption_l_100km = float(fuel_consumption.replace(' l/100 km', '').replace(',', '.'))
        except ValueError:
            return None  # Handle cases where conversion fails
    elif isinstance(fuel_consumption, float):
        fuel_consumption_l_100km = fuel_consumption
    else:
        return None  # If it's neither, return None

    # Check for zero fuel consumption to avoid division by zero
    if fuel_consumption_l_100km == 0:
        return None  # Or you could return a specific value like 0 or -1 to indicate an invalid case

    # Calculate the average range
    average_range = (tank_capacity / fuel_consumption_l_100km) * 100
    return round(average_range, 2)  # Round to 2 decimal places


```

```python
# Code for electric vehicles
    # Average battery capacity in kWh (assuming 60 kWh for this example)
average_battery_capacity_kWh = 60  # kWh

# Function to calculate average range based on efficiency
def calculate_average_range_ev(row):
    # Extract the efficiency value from the row
    efficiency = row['Efficiency']

    # Ensure the efficiency is a float
    if isinstance(efficiency, str):
        # Convert efficiency from string to float (e.g., "173 Wh/km" to 173)
        try:
            efficiency_wh_km = float(efficiency.replace(' Wh/km', '').replace(',', '.'))
        except ValueError:
            return None  # Handle cases where conversion fails
    elif isinstance(efficiency, (float, int)):
        efficiency_wh_km = efficiency
    else:
        return None  # If it's neither, return None

    # Convert efficiency from Wh/km to kWh/km
    efficiency_kwh_per_km = efficiency_wh_km / 1000  # Convert Wh/km to kWh/km

    # Calculate the average range based on battery capacity
    if efficiency_kwh_per_km == 0:
        return None  # Avoid division by zero

    average_range = average_battery_capacity_kWh / efficiency_kwh_per_km  # kWh / (kWh/km)
    return round(average_range, 2)  # Round to 2 decimal places
```

### Generating Code to display the new columns that was created with the total average range for both traditional and electric vehicle.

```python
# Apply the function to calculate the average range for each vehicle
df1_sampled['average_range_norm'] = df1_sampled.apply(calculate_average_range, axis=1)

# Apply the function to calculate the average range for each electric vehicle
df2_sampled['average_range_ev'] = df2_sampled.apply(calculate_average_range_ev, axis=1)
```
